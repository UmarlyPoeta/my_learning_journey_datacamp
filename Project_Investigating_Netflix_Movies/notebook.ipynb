## The data

| Column | Description |
|--------|-------------|

```python
# Importing pandas and matplotlib
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Read in the Netflix CSV as a DataFrame
netflix_df = pd.read_csv("netflix_data.csv")
```

```python
# view dataframe
netflix_df
```

```python
# movie durations
movie_duration = netflix_df[np.logical_and(netflix_df.type == "Movie",netflix_df.release_year>=1990,netflix_df.release_year<2000)].loc[:,"duration"]
```

```python
# distribution of movie duration in the 1990s

plt.hist(movie_duration,100)
plt.show()
```

```python
# What was the most frequent movie duration in the 1990s? Save an approximate answer as an integer called duration.

duration = round(np.mean(movie_duration))
print(f"the most frequent movie duration in the 1990s: {duration}")
```

```python
# A movie is considered short if it is less than 90 minutes. Count the number of short action movies released in the 1990s and save this integer as short_movie_count.

short_movie_counter = 0
for lab, row in netflix_df.iterrows():
    if row["duration"] < 90 and row["genre"]=="Action" and row["release_year"]>=1990 and row["release_year"]<2000:
        short_movie_counter +=1

short_movie_count = short_movie_counter
print(f"number of short action movies released in the 1990s: {short_movie_count}")
```
